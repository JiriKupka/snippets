```python
%matplotlib inline

import random, math
import numpy as np
import cv2
from scipy import spatial
import matplotlib.pyplot as plt
import pickle

def gaussian(x,y,a = 1.0, sigma_x = 2.5, sigma_y = 2.5):
    return a*math.exp(-(x**2/(2*sigma_x**2)+(y**2/(2*sigma_y**2))))


def choose_random_pixels(img, samples):
    width = img.shape[1]
    height = img.shape[0]
    output_samples = []

    for i in range(samples):
        rx = random.randint(0,width-1)
        ry = random.randint(0,height-1)

        output_samples.append(((rx,ry),img[ry][rx]))

    return output_samples
```

```python
def process_image(path, gp = False):
    img = cv2.imread(path,1)
    #img = cv2.resize(img,(img.shape[1]/2,img.shape[0]/2))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)/255.0

    pixel_img = np.zeros(img.shape,np.float32)
    output_img = np.zeros(img.shape,np.float32)

    pixel_density = (img.shape[0]*img.shape[1])/50
    
    if gp:
        img_grey = cv2.imread(path,0)/255.0
        kernel = np.ones((5,5),np.uint8)
        gradient = cv2.morphologyEx(img_grey, cv2.MORPH_GRADIENT, kernel)

        pixels = []
        for y in range(gradient.shape[0]):
            for x in range(gradient.shape[1]):
                if gradient[y][x] > 0.5:
                    pixels.append(((x,y),img[y][x]))
                    
        pixels += choose_random_pixels(img, pixel_density)
    else:
        pixels = choose_random_pixels(img, pixel_density)

    for pixel in pixels:
        pixel_img[pixel[0][1]][pixel[0][0]] = pixel[1]

    tree = spatial.cKDTree([(pixel[0][0],pixel[0][1]) for pixel in pixels])

    for y in range(img.shape[0]):
        for x in range(img.shape[1]):

            points = tree.query_ball_point([x,y],30)

            px_value, factor = 0.0, 0.0

            for point in points:
                dist = gaussian(abs(pixels[point][0][0]-x),abs(pixels[point][0][1]-y))
                px_value += pixels[point][1]*dist
                factor += dist

            if factor > 0:
                output_img[y,x] = px_value/factor

            # point = tree.query([x,y])
            # output_img[y,x] = pixels[point[1]][1]
    return img, pixel_img, output_img, pixels


def plot_images(images):
    fig = plt.figure(figsize=(18,18))
    fig.add_subplot(1,3,1)
    plt.imshow(images[0])
    fig.add_subplot(1,3,2)
    plt.imshow(images[1])
    fig.add_subplot(1,3,3)
    plt.imshow(images[2])
```

```python
images = process_image('dog.png')
pxs = images[0].shape[0]*images[0].shape[1]
ratio = (len(images[3])/float(pxs))*100.0
print "Obrázek zrekonstruován z {0} px, což je pouze {1} % původního obrazu".format(len(images[3]),ratio)
plot_images(images)
```
